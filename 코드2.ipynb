```python
# 박스플롯 값 출력 (이상치 제거된)
q1 = df_salary["salary"].quantile(q=0.25)   #1사분위수
q3 = df_salary["salary"].quantile(q=0.75)   #3사분위수
IQR = q3-q1     #3분위수 -1분위수 =박스플롯 값을 출력
IQR
```

```python
data_IQR = df_salary[(df_salary["salary"] < q3 + IQR * 1.5)&  (df_salary["salary"] > q1 - IQR * 1.5)]
data_IQR     #제3사분위(75%상위값) & 제1사분위(25% 하위값)을 data_IQR에 입력 #박스플롯 위 아래 값까지 출력 
              #IQR: 사분위수 범위 Q3-Q1
```

```python
data_IQR["salary"].hist() #박스플롯 데이터를 히스토그램으로 출력
```

```python
#박스플롯 위에 이상치값을 출력
data_outlier = df_salary[(df_salary["salary"] > q3 + IQR * 1.5)]  
data_outlier
```
